```python
import matplotlib.pyplot as plt
import matplotlib   
import pandas as pd
import seaborn as sns
import os
import numpy as np
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
from mpl_toolkits.mplot3d import Axes3D
```

```python
sns.set() # set seaborn as default style
```

```python
data_dir = os.getcwd()
if 'IN5550' not in data_dir:
    data_dir = os.path.join(data_dir, 'IN5550')
if 'Oblig1' not in data_dir:
    data_dir = os.path.join(data_dir, 'Oblig1')
if 'output' not in data_dir:
    data_dir = os.path.join(data_dir, 'output')
data_dir
```

```python
header = 'act'
filename = os.path.join(data_dir, header+'_4.csv')
data = pd.read_csv(filename, header=0, index_col=0)
data.head()
```

```python
sigmoid = data.loc['sigmoid']
tanh = data.loc['tanh']
relu = data.loc['relu']
```

```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5), sharey=True)
fig.suptitle('3 hidden layer act funcitons vs 4 output layer act. funcitons')

# Sigmoid
sns.barplot(ax=axes[0], x=sigmoid.index, y=sigmoid.values)
axes[0].set_title(sigmoid.name)
axes[0].set(ylim=(1100, 1250))

# tanh
sns.barplot(ax=axes[1], x=tanh.index, y=tanh.values)
axes[1].set_title(tanh.name)

# relu
sns.barplot(ax=axes[2], x=relu.index, y=relu.values)
axes[2].set_title(relu.name)
```

```python
# Static plot titles
htm_titles = [  "Accuracy Score Heatmap",
                "Macro Precision Score Heatmap",
                "Macro Recall Score Heatmap",
                "Macro F1 Score Heatmap",
                "Training elapsed time in seconds"]
def heatmap(header):
    fig, axes = plt.subplots(2, 3, figsize=(20, 10), sharey=False)
    fig.suptitle('All heat maps')
    
    j = 0
    for i, title in enumerate(htm_titles):
        header = 'act'
        filename = os.path.join(data_dir, header+f'_{i}.csv')
        data = pd.read_csv(filename, header=0, index_col=0)
        if i>2:
            i -= 2
            j = 1
        sns.heatmap(
            ax=axes[j][i],
            data=data,
            annot=True,
            fmt=".4g",
            # xticklabels=out_actFcts,
            # yticklabels=hl_actFcts
        )
        axes[j][i].set_title(title)
        
    
    plt.show()
```

```python
heatmap('act')
```
